```python
import pandas as pd 
```

```python
data=pd.read_csv("/home/placement/Downloads/fiat500.csv")
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data1=data.drop(['ID','lat','lon'],axis=1) 
```

```python
data1
```

```python
data=data.loc[(data.model=='lounge')]
```

```python
data
```

```python
data1=pd.get_dummies(data)
```

```python
data1
```

```python
data.shape
```

```python
y=data1['price']
x=data1.drop('price',axis=1)
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_test.head()
```

```python
y_test.head()
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

alpha = [1e-15, 1e-10, 1e-8, 1e-4, 1e-3,1e-2, 1, 5, 10, 20,30]

ridge = Ridge()

parameters = {'alpha': alpha}

ridge_regressor = GridSearchCV(ridge, parameters)

ridge_regressor.fit(x_train, y_train)
```

```python
ridge_regressor.best_params_
```

```python
ypred=ridge_regressor.predict(x_test)
```

```python
ypred
```

```python
ridge=Ridge(alpha=30)
ridge.fit(x_train,y_train)
y_pred_ridge=ridge.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
Ridge_Error=mean_squared_error(y_pred_ridge,y_test)
Ridge_Error
```

```python
from sklearn.metrics import r2_score
r2_score (y_test,y_pred_ridge)
```

```python
Results=pd.DataFrame(columns=['Actual','predicted'])
Results['Actual']=y_test
Results['Predicted']=ypred
Results=Results.reset_index()
Results['Id']=Results.index
Results
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x='Id',y='Actual',data=Results.head(50))
sns.lineplot(x='Id',y='Predicted',data=Results.head(50))
plt.plot()
```
