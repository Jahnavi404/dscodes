```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv("/home/placement/Downloads/TelecomCustomerChurn.csv")
```

```python
data['TotalCharges']=pd.to_numeric(data['TotalCharges'],errors='coerce')
```

```python
data1=data.fillna(data.median())
```

```python
data2=data1.drop(['customerID','customerID','SeniorCitizen','Partner','Dependents','PhoneService','OnlineSecurity','OnlineBackup','DeviceProtection','StreamingTV','StreamingMovies','PaperlessBilling','PaymentMethod',],axis=1)
```

```python
data2
```

```python
data2['Churn']=data2['Churn'].map({'Yes':1,'No':0})
```

```python
data2
```

```python
data3=pd.get_dummies(data2)
```

```python
data3
```

```python
data3.shape
```

```python
y=data3['Churn']
x=data3.drop('Churn',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
